## **IMPORTE**

```python
# Data Handling
import pandas as pd
import numpy as np

# Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv(filepath_or_buffer="../data/train.csv")
test = pd.read_csv(filepath_or_buffer="../data/test.csv")
```

## **INTEGRAÇÃO DE DADOS**

```python
# Verificação de integração dos sets de treino e teste
print(train.shape)
print(test.shape)
print(train.columns.to_list() == test.columns.to_list())
```

```python
# Já foi verificado antes, e não há features com relação temporal
# Permitidno a concatenção e embaralhamento do dataset inteiro
df = pd.concat(objs=[train, test], axis=0, ignore_index=True, copy=True)
df.head()
```

### MATRIZ DE CORRELAÇÃO

$correlacao (x^i, x^j) = \frac{covariância(x^i, x^j)}{s_i * s_j}$

```python
plt.figure(figsize=(13,5))

corr = df.corr(numeric_only=True) # Removido colunas de texto para o cálculo
mask = np.triu(np.ones_like(corr, dtype=np.bool)) # Mascará para visualizar apenas um dos lados simétricos da matriz

sns.heatmap(data=corr, vmin=-1, vmax=1, mask=mask) 
plt.title('Mapa Térmico da Correlação')
```

```python
# Análise focada na correlação dos atributos com o atributo alvo
plt.figure(figsize=(13,2))

mask = np.triu(np.ones_like(corr.loc['target_deathrate':'target_deathrate', :], dtype=np.bool)) # Mascará para visualizar apenas um dos lados simétricos da matriz

sns.heatmap(data=corr.loc['target_deathrate':'target_deathrate', :], vmin=-1, vmax=1) 
plt.title('Mapa Térmico da Correlação')
```
